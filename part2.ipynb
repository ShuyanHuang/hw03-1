```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")

# Load the example mpg dataset
mpg = sns.load_dataset("mpg")

# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(40, 400), alpha=.5, palette="muted", 
            height=6, data=mpg)
sns.regplot(x="horsepower", y="mpg", data=mpg, 
           order=2, ci=None, scatter=False)
plt.title('Fuel Economy, Horsepower and Weight of Automobiles from Different Countries')
plt.show()
```

This figure shows the relationship among vehicles' fuel economy, horsepower and weight, together with the place of production of the vehicles. I use the mpg dataset from the StatLib library which is maintained at Carnegie Mellon University. By this data visualization, I aim to test the common impression that vehicles with larger horse power are heavier and more fuel-consuming. Also, I want to know whether there's any country specific feature in vehicle's fuel economy, horsepower and weight.

I choose to use a scatter plot mainly because by plotting two continuous variables on the x and y axes it's straightforward to show the relationship between the two variables, and by representing each sample with a dot, I can also explore individual behaviors. Besides, by using different colors and sizes of the dots, more variables can be included in the plot, without making it more complicated. Since the major variables that I am interested in are mpg and horsepower, I use them as the x and y axes, respectively. It's very intuitive to encode the weight of the vehicles as the size of the dots, since heavier vehicles "feel" larger. It's also straightforward to use different colors for vehicles produced in different countries since the place of production is a categorical variable and there are only three possible values. In addition to the scatter plot, I also draw the fitted line of a 2-order polynomial regression of variables horsepower and mpg, to give a more clear demonstration of the relationship between the two variables.

From the scatter plot and the fitted regression line, we can see a generally negative relationship between mpg and horsepower. This makes intuitive sense since vehicles with more powerful engines usually require more energy. However, it can be clearly seen that this relationship is not linear. The fitted line flattens and then goes up a little at the right tail of the horsepower axis. Therefore, there's no significant negative relationship between mpg and horsepower among the vehicles with horsepower larger than 150. Some vehicles with extremely large horsepower may have moderate mpg. Also, there are some other outliers worth noticing. One vehicle produced in Japan has horsepower around 125 and mpg around 33, which is much higher than the mpg of vehicles with comparable horsepower, so this vehicle is less energy-efficient than average level. A vehicle preoduced in the U.S. has horsepower around 75 and mpg around 15, which is much lower than the mpg of vehicles with comparable horsepower, indicating that it's more fuel-saving than average. 

As for the patterns in weight, in general, more powerful vehicles are heavier, since the dots in the right region of the figure are larger. And U.S. vehicles are usually heavier than comparable Japanese and European vehicles. In terms of places of production, U.S. produces more vehicles with large horsepower and low mpg, which can be easily seen from the dominating blue dots in the right region of the figure.
